# For loop

```python
# Print numbers from 1 to 10
for i in range(1,11):
  print(i)
```

```python
# Print reverse number from 10 to 1
for i in range(10,0,-1):
  print(i)
```

```python
# Print even and odd numbers between 1 to 100
for j in range(2, 101, 2):
  print("Even:",j)
for i in range(1, 101, 2):
  print("ODD:",i)
```

```python
# Multiplication table of 5
for i in range(1,11):
  print(5*i)
```

```python
# Factorial of a number
n = int(input("Enter a number:"))
print(type(n))
result = 1
for i in range(1, n+1):
  result = result * i
print(result)
```

```python
# Prime between 1 to 100
result = set()
for i in range(2,101):
  for j in range(2,i):
    if i % j == 0:
      break
  else:
    result.add(i)
print(result)
```

```python
# check the prime number
num = int(input("Enter a number:"))

if num <= 1:
    print("Not Prime")
else:
    for i in range(2, num):
        if num % i == 0:
            print("Not Prime")
            break
    else:
        print("Prime")
```

```python
# Print fibinocci series
n = int(input("Enter how many terms: "))

a, b = 0, 1

for i in range(n):
    print(a)
    a, b = b, a + b
```

```python
#Factors
sum_of_factors = 0
num = int(input("Enter a number: "))

for i in range(1, num + 1):
    if num % i == 0:
        print("factors:",i)
        sum_of_factors += i

print("sum of factors:" ,sum_of_factors)
```

```python
# hcm and lcm

a = int(input("Enter first number: "))
b = int(input("Enter second number: "))

hcf = 1
for i in range(1, min(a, b) + 1):
    if a % i == 0 and b % i == 0:
        hcf = i


lcm = (a * b) // hcf

print("LCM:", lcm)
print("HCF:", hcf)
```

```python
# squares and cube from 1 to n
n = int(input("Enter a number: "))

for i in range(1, n +1):
  print(i," -sq-> ",i*i)
  print(i, " -cu-> ", i * i * i)

```

```python
# sum of even numbers and odd numbers
n = int(input("Enter a number: "))
even , odd = 0,0
for i in range(1, n +1):
  if i % 2 == 0:
    print(i," even")
    even += i
  else:
    print(i," odd")
    odd += i

print("sum of odd numbers:",odd)
print("sum of even numbers:",even)
```

# While loop

```python

# Print all numbers from 1 to 10 using a loop.

i = 1

while i < 11:
    print(i)
    i += 1
```

```python
# Print numbers from 10 down to 1 in reverse order.
i = 10

while i > 0:
    print(i)
    i -= 1
```
