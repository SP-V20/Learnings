```python
# Print numbers from 1 to 10
for i in range(1,11):
  print(i)
```

```python
# Print reverse number from 10 to 1
for i in range(10,0,-1):
  print(i)
```

```python
# Print even and odd numbers between 1 to 100
for j in range(2, 101, 2):
  print("Even:",j)
for i in range(1, 101, 2):
  print("ODD:",i)
```

```python
# Multiplication table of 5
for i in range(1,11):
  print(5*i)
```

```python
# Factorial of a number
n = int(input("Enter a number:"))
print(type(n))
result = 1
for i in range(1, n+1):
  result = result * i
print(result)
```

```python
# Prime between 1 to 100
result = set()
for i in range(2,101):
  for j in range(2,i):
    if i % j == 0:
      break
  else:
    result.add(i)
print(result)
```
